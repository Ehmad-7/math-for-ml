```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
tosses=np.random.randint(0,2,size=1000)

tosses_series=pd.Series(tosses).map({0:"Heads",1:"Tails"})
counts=tosses_series.value_counts()


print("Coin tosses results for 1000 tossess:")
print(counts)
```

```python
counts.plot(kind='bar',color=['skyblue','salmon'],edgecolor="black")
plt.title("Coin Toss Simulation (1000 trials)")
plt.xlabel("Outcome")
plt.ylabel("Frequency")
plt.show()
```

## Probablity Distribution

```python
running_heads_prob=(tosses_series=="Heads").expanding().mean()
plt.plot(running_heads_prob,color="purple")
plt.axhline(0.5,color="red",linestyle="--",label="Expected Probability = 0.5")
plt.title("Running Probability of Heads")
plt.xlabel("Number of Tosses")
plt.ylabel("Probability")
plt.legend()
plt.show()
```
