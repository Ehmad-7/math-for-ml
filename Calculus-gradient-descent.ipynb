## Gradient Descent

```python
import matplotlib.pyplot as plt
```

```python
# function and derivative

def f(x):
  return x**2

def f_prime(x):
  return 2*x

x=10
learning_rate=0.1
iterations=25

history=[x]

for i in range(iterations):
  grad=f_prime(x)
  x=x-learning_rate*grad
  history.append(x)

print(f"Minimum at x ≈ {x},f(x) ≈ {f(x)}")
```

```python
xs=range(len(history))
plt.plot(xs,[f(h) for h in history],marker='o')
plt.title("Gradient Descent on f(x) = x^2")
plt.xlabel("Iteration")
plt.ylabel("f(x)")
plt.show()
```
